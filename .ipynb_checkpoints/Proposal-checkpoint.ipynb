Introduction: 

Pulsars are rare and powerful neuron stars that have unique radio wave emission patterns and are searched by observing periodic radio signals. Pulsars are researched for their abilities as probes of space-time, the inter-stellar medium, and states of matter. 

Finding genuine pulsar signals are difficult due to most detections being radio frequency interference (RFI) and noise (Lyon et. al., 2016). As such, we will be addressing this by creating a classification model to predict whether unknown data points are pulsars. The dataset we are working with was collected by Lyon et. al., and contains thousands of RFI/noise and true pulsar emission examples, with the latter being the positive minority.


Preliminary exploratory data analysis:

The dataset can be read into R by first uploading the data file into R, and by using the tidyverse package in order to be able to perform classification and visualization.

```r
library(tidyverse)
library(repr)
options(repr.matrix.max.rows = 6)
```

```r
pulsar_data <- read_csv("data/HTRU_2.csv")
```

To clean the data we renamed the columns as follows:

```r
colnames(pulsar_data) <- c('mean_profile', 'std_profile', 
                           'kurtosis_profile', 'skewness_profile', 
                           'mean_dmsnr', 'std_dmsnr', 'kurtosis_dmsnr', 
                           'skewness_dmsnr', "is_pulsar")
```

```r
pulsar_data
```

In order to have tidy data, we also converted is_pulsars to be a factor data type in the next code using mutate and as_factor function.

We will be using the following table for our exploratory data analysis.

```r
select_pulsar_data<-pulsar_data|>
    select(mean_profile, skewness_profile,is_pulsar)|>
    mutate(is_pulsar=as_factor(is_pulsar))

select_pulsar_data
```

```r
options(repr.plot.width=10,repr.plot.length=10)
plot_pulsar_data<-ggplot(select_pulsar_data, aes(x=mean_profile, y=skewness_profile))+
                         geom_point(aes(color=is_pulsar))+
                         labs(x="mean of the integrated profile", y="skewness the integrated profile", color="Status of if the integrated profile is a pulsar, 1=yes, 0=no")
            
plot_pulsar_data    
```

Methods:

Our data analysis will focus on three out of the nine attributes in the data, specifically the mean and skewness of the integrated profile (GHz), and the factor variable, of whether it is a pulsar (=1) or not (=0). The integrated profile shows the distinct identity of each individual pulsar and remains stable throughout time (Liu, 2017). 

Using the KKNN algorithm, our goal is to develop a classification model that can determine whether or not a particular point may be categorised as a pulsar. We will randomly split the available data into training and testing sets. We will then perform cross-validation to assess the model and ensure it does not overfit the data.

To visualise this data, a scatterplot is used. Skewness of the integrated profile on the y-axis, with the mean of the integrated profile on the x-axis. The is_pulsar class is shown on the scatterplot through the different colours of data points; blue observations are verified pulsars and orange points are verified non-pulsars.


Expected outcomes and significance:

Our classification model will give the expected output of predicted pulsar status and can be used by others as a tool to prioritize their efforts to first research predicted pulsar points. Further questions should be considered such as how can we improve the accuracy of our model considering how to best address any missing datapoints and outliers? Are there more tale-tell signs of pulsars that we should consider apart from difficult to measure radio-wave frequencies? Moreoever, are there other attributes that are included in the dataset that may be better for classifying pulsars than the mean and skewness of the integrated profile? For example, excess kurtosis of the integrated profile, or use statistics of DM-SNR curve estimates.
